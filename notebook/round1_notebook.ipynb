```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_1=pd.read_csv("../prices/round1/prices_round_1_day_-2.csv", sep=";")
df_2=pd.read_csv("../prices/round1/prices_round_1_day_-1.csv", sep=";")
df_3=pd.read_csv("../prices/round1/prices_round_1_day_0.csv", sep=";")
```

```python
#PRODUCT SELECTED
# product="INTARIAN_PEPPER_ROOT"
product="ASH_COATED_OSMIUM"
product_prices_1=df_1[df_1["product"]==product]
product_prices_1=product_prices_1[["timestamp","mid_price"]]

product_prices_2=df_2[df_2["product"]==product]
product_prices_2=product_prices_2[["timestamp","mid_price"]]

product_prices_3=df_3[df_3["product"]==product]
product_prices_3=product_prices_3[["timestamp","mid_price"]]
```

```python
plt.plot(product_prices_1["timestamp"],product_prices_1["mid_price"])
plt.show()
```

```python
plt.plot(product_prices_2["timestamp"],product_prices_2["mid_price"])
plt.show()
```

```python
plt.plot(product_prices_3["timestamp"],product_prices_3["mid_price"])
plt.show()
```

```python
product_prices_2["timestamp"]=product_prices_2["timestamp"]+product_prices_1["timestamp"].iloc[-1]+100
product_prices_3["timestamp"]=product_prices_3["timestamp"]+product_prices_2["timestamp"].iloc[-1]+100
product_prices=pd.concat([product_prices_1,product_prices_2,product_prices_3])
end=300
plt.plot(product_prices["timestamp"][:end],product_prices["mid_price"][:end])
plt.show()
```
