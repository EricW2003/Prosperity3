```python
import numpy as np
```

```python
bid_list=np.array([0,0,0,30,5,12,28])
ask_list=np.array([30,20,20,0,0,40,0])
prices=np.array([33,32,31,30,29,28,27])
N=len(bid_list)
```

```python
def calculate_price_volumes(bid_list,ask_list):
    cum_bid=np.cumsum(bid_list)
    cum_ask=np.cumsum(ask_list[::-1])
    ans=np.minimum(cum_bid,cum_ask[::-1])
    return np.argmax(ans),np.max(ans)

print(calculate_price_volumes(bid_list,ask_list))
```

```python
def calculate_gains(price_ind,volume, side):
    bid=bid_list.copy()
    ask=ask_list.copy()
    if side=="BUY":
        bid[price_ind]+=volume
    else:
        ask[price_ind]+=volume
    cl_price_ind,cl_volume=calculate_price_volumes(bid,ask)
    if side=="BUY":
        if price_ind>cl_price_ind:
            return 0
        lost_vol=np.cumsum(bid_list)[price_ind]
    else:
        if price_ind<cl_price_ind:
            return 0
        lost_vol=np.cumsum(ask_list[::-1])[::-1][price_ind]
    vol_accessible=max(0,cl_volume-lost_vol)

    ans=min(volume,vol_accessible)*(30-prices[cl_price_ind])
    return ans if side=="BUY" else -ans
```

```python
side="BUY"
vol_max=110000
max_gain=0
opt_price=None
opt_vol=None
for price_ind in range(len(prices)):
    price_ind=len(prices)-1-price_ind
    for volume in range(1,vol_max+1):
        gain=calculate_gains(price_ind,volume/1000, side)
        if gain>max_gain:
            max_gain=gain
            opt_price=prices[price_ind]
            opt_vol=volume/1000
print(max_gain,opt_price,opt_vol)
```
