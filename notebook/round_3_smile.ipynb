```python
import numpy as np
import random as rd
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
implied_volatilities = pd.read_csv("../notebook/implied_volatilities.csv", sep=",")
underlying = implied_volatilities["VELVETFRUIT_EXTRACT"].to_numpy()
```

```python
strikes = [6500,5200,5500,4500,5300,5400,5100,6000,5000,4000]
# strikes = [4000]
colors = [
    "blue","red","green","orange","purple",
    "brown","pink","gray","olive","cyan"
]

for i, strike in enumerate(strikes):
    plt.scatter(
        underlying / strike,
        implied_volatilities[f"{strike}"].to_numpy(),
        color=colors[i],
        label=f"VENV_{strike}"
    )

plt.legend()
plt.title("SMILE")
plt.xlabel("Moneyness")
plt.ylabel("Implied Volatilities")
plt.show()
```
