```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_1=pd.read_csv("./csv/prices/prices_round_0_day_-2.csv", sep=";")
df_1.head()
```

```python
df_2=pd.read_csv("./csv/prices/prices_round_0_day_-1.csv", sep=";")
df_2.head()
```

```python
product="EMERALDS"
# product="TOMATOES"
product_prices_1=df_1[df_1["product"]==product]
product_prices_1=product_prices_1[["timestamp","mid_price"]]

product_prices_2=df_2[df_2["product"]==product]
product_prices_2=product_prices_2[["timestamp","mid_price"]]


```

```python
plt.plot(product_prices_1["timestamp"],product_prices_1["mid_price"])
plt.show()
```

```python
plt.plot(product_prices_2["timestamp"],product_prices_2["mid_price"])
plt.show()
```

```python
product_prices_2["timestamp"]=product_prices_2["timestamp"]+product_prices_1["timestamp"].iloc[-1]+100
product_prices_2.head()
product_prices=pd.concat([product_prices_1,product_prices_2])

plt.plot(product_prices["timestamp"][:7],product_prices["mid_price"][:7])
plt.show()
```
